```python
#Coloumn contains values like 'Younger than 5 years' and 'Older than 85' not valid input to perform analysis
# Replacing 'Younger than 5 years' with 4 and Older than 85 with 86
#inplace will change current dataframe  and if inplace=False then dataframe will not change 
df['Age1stCode'].replace({'Younger than 5 years' : '4', 'Older than 85' : '86'} , inplace=True)
```

```python
#Deduceing the percentage of developers who know python in each country.
```

```python
#Numbers of people responsed in each country
response = df['Country'].value_counts()
response
```

```python
#Now grouping responders based on their country
filt = df.groupby(['Country'])
filt
```

```python
#Now filtering people who worked with python
people_know_python = filt['LanguageWorkedWith'].apply((lambda x: x.str.contains('Python').sum()))
```

```python
people_know_python.head(50)
```

```python
#now concating country and people who worked with python
python_df = pd.concat([response , people_know_python] , axis='columns' , sort=False)
python_df
```

```python
#now renameing 'country' to 'devlopers' and 'LanguageWorkedWith' to 'python_users'
python_df.rename(columns = {'Country' : 'devlopers' ,'LanguageWorkedWith' : 'python_users' }, inplace=True)
python_df
```

```python
#percentage of people who knows python based on their country = (country/people_who_knows_python)*100 
python_df['pctknowspython'] = (python_df['python_users'] / python_df['devlopers']  ) * 100
python_df
```
